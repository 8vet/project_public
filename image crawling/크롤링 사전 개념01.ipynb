## urllib.parse - url을 구성요소로 구문 분석
1. url 구문 분석 :
urllib.parse.urlparse(urlstring, scheme='', allow_fragments=True)

2. urljoin

>> from urllib.parse import urljoin
>> urljoin('http://www.cwi.nl/%7Eguido/Python.html', 'FAQ.html')
출력 'http://www.cwi.nl/%7Eguido/FAQ.html'

3. quote

%xx 이스케이프를 사용하여 string 의 특수 문자를 치환
>> quote('/El Niño/')는 '/El%20Ni%C3%B1o/'를 산출

4. quote_plus

quote()와 유사하지만 스페이스를 더하기 부호로 치환하기도 합니다
>> quote_plus('/El Niño/')는 '%2FEl+Ni%C3%B1o%2F'를 산출

참고: https://docs.python.org/

## GET방식 POST방식 
GET 방식:
? 뒤에 주소값이 붙는것
>>  예시 : http://url/bbslist.html?id=5&pagenum=2 

POST 방식:
form을 이용해서 submit 을 하는 형태

언제 써야할지) 쉽게 생각하면, get 은 가져오는것(select), post는 수행하는것 (서버의 값이나 상태를 바꾸기 위해서 사용하는것처럼,..)

```python
from requests.compat import urlparse, urljoin
from requests.exceptions import HTTPError
from urllib.robotparser import RobotFileParser
import time
```

```python
def canfetch(url, agent='*', path='/'):
    robot = RobotFileParser(urljoin(url,'/robots.txt'))
    robot.read()
    return robot.can_fetch(agent,urlparse(url)[2])

def download(url,params={}, headers={},method='GET',limit=3):
    if canfetch(url)==False:
        print('[ERROR]'+url)
    
    try:
        resp = requests.request(method, url,
                params=params if method=='GET' else {},
                data = params if method=='POST' else {},
                headers = headers)
        resp.raise_for_status()
    except HTTPError as e:
        if limit>0 and e.response.status_code >= 500:
            print(limit)
            time.sleep(1)
            resp = download(url,params, headers, method, limit-1)
        else :
            print('[{}] '.format(e.response.status_code) + url)
            print(e.response.status_code)
            print(e.response.reason)
            print(e.response.headers)
    return resp
```

```python
url = 'https://www.google.com/search'
params ={
    'q':'',
    'oq':'',
    'aqs':'hrome..69i57j69i59j69i65l3j69i61j69i60j69i61.1205j0j7',
    'sourceid':'chrome',
    'ie':'UTF-8'
}
params['q']=params['oq']='파 이 썬'
headers = {'user-agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36'
}
resp = download(url, params,headers, 'GET')
resp.text
```
